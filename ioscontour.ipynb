```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x = np.linspace(-3,3,100)
y = np.linspace(-3,3,100)
x,y= np.meshgrid(x,y)
z_functions= [
    np.sin(x)+np.cos(y),
    np.sin(x)*np.cos(y),
    np.exp(-x**2 - y**2),
    np.sqrt(x**2 + y**2),
    np.tan(x)+ np.tan(y),
    x**2 - y**2
]

z_functions_name =[
    r'$\sin(x) + \cos(y)$',
    r'$\sin(x) \cdot \cos(y)$',
    r'$\exp(-x^2 - y^2)$',
    r'$\sqrt(x^2 + y^2)$',
    r'$\tan(x) +\tan(y)$',
    r'$x^2 -y^2$'
]

cmaps =['viridis', 'plasma',
       'inferno','magma','cividis','twilight']

plt.figure(figsize=(12,8))
for i in range (6):
 plt.subplot(2,3,i+1)
 contour = plt.contour(x,y,z_functions[i], cmap=cmaps[i])
 plt.contour(contour)
 plt.title('function: ()'.format(z_functions_name[i]))
 plt.xlabel(r'$x$')
 plt.ylabel(r'$y$')
 plt.grid(True)
plt.savefig('i.jpg', dpi = 600)
plt.tight_layout()
plt.show()
```
