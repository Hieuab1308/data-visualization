```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,10,100)
y1= np.sin(x)
y2 = np.cos(x)
fig1, axes1 = plt.subplots(nrows=1, ncols = 2, figsize= (12,3))

axes1[0].plot(x, y1, 'r')
axes1[0].set_title('Sin Function')
axes1[1].plot(x,y2,'b')
axes1[1].set_title('cosine function')

plt.tight_layout()
plt.savefig('subplot.jpg', dpi = 600)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA

iris = load_iris()
itis = load_iris
X= iris.data
y = iris.target

pca = PCA(n_components=3)
X_pca = pca.fit_transform(X)

fig = plt.figure(figsize = (8,6))
ax = fig.add_subplot(111, projection = '3d')

ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2], c=y, cmap = 'viridis', s = 50)
ax.set_xlabel('Principal Compoment 1')
ax.set_ylabel('Principal Compoment 2')
ax.set_zlabel('Principal Compoment 3')
ax.set_title('PCA Visalization of Iris Dataset')

plt.savefig('Pca.jpg', dpi = 600)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#x = np.random.uniform(low = -10, high = 10, size = 1000)
#y= np.random.uniform(low = -10, high = 10, size = 1000)
#z = np.random.uniform(low = -0.1, high = 0.1, size = 1000)
x = np.random.normal(0,1, size = 1000)
y = np.random.normal(0,1, size = 1000)
z = np.random.normal(0,1, size = 1000)

fig = plt.figure(figsize= (10,10))
ax= fig.add_subplot(111, projection = '3d')
sc = ax.scatter(x,y,z, c = z, cmap='coolwarm', marker ='o')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z(thickness)')
ax.set_title('Semiconction Wafer')

fig.colorbar(sc, label =' thickness variation')
plt.savefig('es.jpg', dpi = 600)
plt.show()
```

```python
import plotly.graph_objects as go
import numpy as np
#!pip install -U kaleido

x = np.linspace(-5,5,100)
y = np.linspace(-5, 5, 100)
x,y = np.meshgrid(x,y)
z = np.sin(np.sqrt(x**2 +y**2))

fig = go.Figure(data=[go.Surface(z=z,x=x,y=y, colorscale ='jet')])

fig.update_layout(title ='Surface Plot',
                  scene= dict(xaxis_title ='X',
                              yaxis_title ='Y',
                              zaxis_title ='Z'))

#fig.write_image('3d_surface.pdf')
fig.show()
```
