```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

var_x = np.random.normal(0,1,300)
var_y = np.random.uniform(0,1,300)

var_z= var_x + var_y

df = pd.DataFrame([])
df["var x"] = var_x
df["var y"] = var_y
df["var z"] = var_z

plt.figure(figsize=(12,4))
sns.scatterplot(data = df, x= 'var x', y = 'var y')
plt.title('Scatter plot')
plt.xlabel('var_x')
plt.ylabel('var_y')
plt.xticks(rotation = 45)
plt.savefig('scatterplot.jpg',dpi = 600)
plt.show()

plt.figure(figsize=(12,4))
sns.scatterplot(data = df, x= 'var y', y = 'var z')
plt.title('Scatter plot')
plt.xlabel('var_y')
plt.ylabel('var_z')
plt.xticks(rotation = 45)
plt.savefig('scatterplot1.jpg',dpi = 600)
plt.show()

plt.figure(figsize=(12,4))
sns.scatterplot(data = df, x= 'var x', y = 'var z')
plt.title('Scatter plot')
plt.xlabel('var_x')
plt.ylabel('var_z')
plt.xticks(rotation = 45)
plt.savefig('scatterplot2.jpg',dpi = 600)
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

var_x = np.random.normal(0,1,300)
var_y = np.random.uniform(0,1,300)

var_z= var_x + var_y

df = pd.DataFrame([])
df["var x"] = var_x
df["var y"] = var_y
df["var z"] = var_z

nam=["var x", "var y"]
yeu=["var x","var z"]
nhi=["var y", "var z"]
plt.figure(figsize=(12,4))
sns.boxplot(data = df[nam])
plt.title('Box plot')
plt.xlabel('Variables')
plt.ylabel('Values')
plt.xticks(rotation = 45)
plt.savefig('box_plot.jpg',dpi = 600)
plt.show()

plt.figure(figsize=(12,4))
sns.boxplot(data = df[yeu])
plt.title('Box plot')
plt.xlabel('Variables')
plt.ylabel('Values')
plt.xticks(rotation = 45)
plt.savefig('box_plot1.jpg',dpi = 600)
plt.show()

plt.figure(figsize=(12,4))
sns.boxplot(data = df[nhi])
plt.title('Box plot')
plt.xlabel('Variables')
plt.ylabel('Values')
plt.xticks(rotation = 45)
plt.savefig('box_plot2.jpg',dpi = 600)
plt.show()

plt.figure(figsize=(12,4))
sns.histplot(data=df, bins=20, kde=True, color='skyblue')
plt.title('Histogram of sepal length in iris dataset')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.grid(True)
plt.savefig('df_histogram_kde.jpg', dpi = 600)
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

var_x = np.random.normal(0,1,300)
var_y = np.random.uniform(0,1,300)
var_d = var_x**2 + var_y**2 + var_x**2
var_z= var_x + var_y

df = pd.DataFrame([])
df["var x"] = var_x
df["var y"] = var_y
df["var z"] = var_z
df["var d"] = var_d

plt.figure(figsize=(12,4))
sns.scatterplot(data = df, x = 'var d', y = 'var x')
plt.title('Scatter plot')
plt.xlabel('var_d')
plt.ylabel('var_x')
plt.xticks(rotation = 45)
plt.savefig('scatterplotDX.jpg',dpi = 600)
plt.show()


plt.figure(figsize=(12,4))
sns.scatterplot(data = df, x = 'var d', y = 'var y')
plt.title('Scatter plot')
plt.xlabel('var_d')
plt.ylabel('var_y')
plt.xticks(rotation = 45)
plt.savefig('scatterplotDY.jpg',dpi = 600)
plt.show()


plt.figure(figsize=(12,4))
sns.scatterplot(data = df, x = 'var d', y = 'var z')
plt.title('Scatter plot')
plt.xlabel('var_d')
plt.ylabel('var_z')
plt.xticks(rotation = 45)
plt.savefig('scatterplotDz.jpg',dpi = 600)
plt.show()

plt.figure(figsize=(12,4))
sns.histplot(data=df, bins=20, kde=True, color='skyblue')
plt.title('Histogram of sepal length in iris dataset')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.grid(True)
plt.savefig('df_histogram_kdeD.jpg', dpi = 600)
plt.show()
```
